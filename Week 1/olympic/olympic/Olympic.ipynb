```python
medal_counts = df.groupby(['Year', 'Gender']).size().unstack()

fig_gen, ax_gen = plt.subplots(figsize=(10, 6), facecolor='#EAF4f4')

ax_gen.set_facecolor('#EAF4f4')

ax_gen.plot(medal_counts.index, medal_counts['Men'], color='#005f73', label='Men', marker='o')
ax_gen.plot(medal_counts.index, medal_counts['Women'], color='#94d2bd', label='Women', marker='o')

ax_gen.fill_between(medal_counts.index, medal_counts['Men'], medal_counts['Women'], color='#005f73', alpha=0.6)
ax_gen.fill_between(medal_counts.index, 0, medal_counts['Women'], color='#94d2bd', alpha=0.6)

ax_gen.set_ylabel('Number of Won Medals', color='#22333b', fontsize=10)
ax_gen.set_title('Male and Female Won Medals over the Years', color='#22333b', fontsize=12)
ax_gen.tick_params(axis='both', length=0, colors='#22333b')

legend = ax_gen.legend(facecolor='#EAF4f4')


# Set the y-axis on the right side
ax_gen.yaxis.tick_right()

# Set the x-axis limit
ax_gen.set_xlim(1896, 2012)

# Remove the spines
ax_gen.spines['right'].set_visible(False)
ax_gen.spines['top'].set_visible(False)
ax_gen.spines['left'].set_visible(False)
```

```python
medals_by_discipline = pd.DataFrame({
    'Sport': ['Aquatics', 'Canoe', 'Canoe / Kayak', 'Judo', 'Shooting', 'Wrestling'],
    'Bronze': [0.0, 3.0, 5.0, 0.0, 3.0, 1.0],
    'Gold': [0.0, 0.0, 10.0, 0.0, 0.0, 0.0],
    'Silver': [2.0, 0.0, 7.0, 1.0, 2.0, 0.0]
}).set_index('Sport')

# Create the outer donut chart
fig_SVK, ax_SVK = plt.subplots(figsize=(10, 8))

fig_SVK.set_facecolor('#EAF4f4')

ax_SVK.pie(medals_by_discipline.sum(axis=1), labels=medals_by_discipline.index,
           colors=['#d8f3dc', '#b7e4c7', '#95d5b2', '#74c69d', '#52b788', '#40916c'],
           radius=1, wedgeprops=dict(width=0.3, edgecolor='w'), startangle=90,
           counterclock=False, textprops=dict(color='#22333b', fontsize=12))

# Calculate the total medals
medals=[ 2.,  3.,  5., 10.,  7.,  1.,  3.,  2.,  1.]
color = ['#E0E1DD'  ,  '#CD7F32'  ,  '#CD7F32'  ,  '#E9C46A'  ,  '#E0E1DD'  ,  '#E0E1DD'  ,  '#CD7F32'  ,  '#E0E1DD'  ,  '#CD7F32'  ]
lable = ['Silver','Bronze','Bronze','Gold', 'Silver' ,'Silver' ,'Bronze', 'Silver' ,'Bronze']


# Create the inner pie chart
ax_SVK = plt.gcf().gca()
ax_SVK.pie(medals, colors=color,labels=lable,  radius=0.7, wedgeprops=dict(width=0.3, edgecolor='w'),
             startangle=90, counterclock=False, 
             pctdistance=0.75, textprops=dict(color='#22333b', fontsize=12))

# Set chart title
ax_SVK.set_title('Count of Medals in each Sport', color='#22333b', fontsize=12)

plt.show()
```

```python
import zlib
import zipfile
import joblib

joblib.dump(fig_gen, "fig_gen")
joblib.dump(fig_top10, "fig_top10")
joblib.dump(fig_SVK, "fig_SVK")

def compress(file_names):
    print("File Paths:")
    print(file_names)
    compression = zipfile.ZIP_DEFLATED
    with zipfile.ZipFile("result.zip", mode="w") as zf:
        for file_name in file_names:
            zf.write('./' + file_name, file_name, compress_type=compression)

file_names = ["fig_gen", "fig_top10", "fig_SVK", "Olympic.ipynb"]
compress(file_names)
```
