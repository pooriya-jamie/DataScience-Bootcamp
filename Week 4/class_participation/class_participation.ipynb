```python
num_samples = 1000
sample_size = 30

samples = np.random.choice(data['ratio'], size=(num_samples, sample_size), replace=True)

samples[:2]
```

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
میانگین و انحراف از معیار هر ست از نمونه‌ها را محاسبه کرده و در متغیر‌های
<code>mean_sample_set</code>
و
<code>std_sample_set</code>
ذخیره کنید.
</font>
</p>

```python
mean_sample_set = np.mean(samples, axis=1)
std_sample_set = np.std(samples, axis=1)

mean_sample_set[:5], std_sample_set[:5]
```

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
 هیستوگرام توزیع نمونه‌‌ها در ۱۰ نمونه‌برداری اول را در قالب ۱۰ تصویر رسم نمایید.

```python
num_samplings_to_display = 10

for i in range(num_samplings_to_display):
    plt.hist(samples[i], bins=15, edgecolor='black', color='green')
    plt.xlabel("Engagement Ratio")
    plt.ylabel("Frequency")
    plt.title(f"Distribution of Samples (Sampling {i+1})")
    plt.show()
```

</font>

```python
plt.hist(mean_sample_set, bins=30, edgecolor='black', color='green')
plt.xlabel("Average Engagement Ratio")
plt.ylabel("Frequency")
plt.title("Distribution of Average Engagement Ratios")
plt.show()
```



<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
تقریبا ۹۵ درصد میانگین نمونه‌ها در چه بازه‌ای قرار می‌گیرد؟ (توجه: این تقریب را بر اساس نرمال فرض کردن توزیع این نمونه‌ها درنظر می‌گیریم.)</font>
</p>

```python
std_error = np.std(mean_sample_set, ddof=1) / np.sqrt(len(mean_sample_set))

# Calculate the margin of error for a 95% confidence interval (z-score for 95% confidence level is approximately 1.96)
margin_of_error = 1.96 * std_error

# Calculate the confidence interval
lower_bound = np.mean(mean_sample_set) - margin_of_error
upper_bound = np.mean(mean_sample_set) + margin_of_error

lower_bound, upper_bound
```


<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
بازه‌ی اطمینان ۹۵ درصد در قسمت قبل را با استفاده از z-table محاسبه کنید.
</font>
</p>


```python
# Calculate the critical value from the z-table for 95% confidence interval
critical_value = norm.ppf(0.975)  # For a two-tailed distribution, we use alpha/2 = 0.025

# Calculate the margin of error
margin_of_error = critical_value * std_error

# Calculate the confidence interval
lower_bound = np.mean(mean_sample_set) - margin_of_error
upper_bound = np.mean(mean_sample_set) + margin_of_error

lower_bound, upper_bound
```

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
برای هر ۱۰۰۰ نمونه‌گیری صورت گرفته در بالا(بخش اول)، بازه‌ی اطمینان ۹۵ درصد را حساب کنید. درصد تعداد نمونه‌هایی که میانگین جمعیت در بازه‌ی اطمینان ۹۵ درصد آن‌ها(نمونه‌ها) در بازه‌ی بدست آمده بوده‌اند را گزارش کنید.
</font>
</p>


```python
confidence_intervals_lower = mean_sample_set - margin_of_error
confidence_intervals_upper = mean_sample_set + margin_of_error

# Check if the population mean is within the confidence interval for each sample
within_interval = np.logical_and(confidence_intervals_lower <= np.mean(data["ratio"]), np.mean(data["ratio"]) <= confidence_intervals_upper)

# Calculate the percentage of samples with population mean in the confidence interval
percentage_within_interval = np.sum(within_interval) / num_samples * 100

percentage_within_interval
```

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
تمامی خواسته‌های بخش اول را این بار با نمونه‌هایی با اندازه ۲۰۰ سمپل تکرار کنید و نتایج را مقایسه کنید.
</font>
</p>

```python
# Calculate the mean for each sample set
sample_size_new = 200
num_samples_new = 1000

samples_new = np.random.choice(data['ratio'], size=(num_samples_new, sample_size_new), replace=True)
mean_sample_set_new = np.mean(samples_new, axis=1)

# Calculate the standard deviation of the sample averages
std_dev_sample_averages_new = np.std(mean_sample_set_new, ddof=1)

# Calculate the critical value from the z-table for 95% confidence interval
critical_value = norm.ppf(0.975)  # For a two-tailed distribution, we use alpha/2 = 0.025

# Calculate the margin of error
margin_of_error_new = critical_value * (std_dev_sample_averages_new / np.sqrt(sample_size_new))

# Calculate the confidence intervals for each sample
confidence_intervals_lower_new = mean_sample_set_new - margin_of_error_new
confidence_intervals_upper_new = mean_sample_set_new + margin_of_error_new

# Check if the population mean is within the confidence interval for each sample
within_interval_new = np.logical_and(confidence_intervals_lower_new <= np.mean(data["ratio"]), np.mean(data["ratio"]) <= confidence_intervals_upper_new)

# Calculate the percentage of samples with population mean in the confidence interval
percentage_within_interval_new = np.sum(within_interval_new) / num_samples_new * 100

percentage_within_interval_new
```

اگر بخواهیم این درس را برای همه دانشجویان این جمعیت آماده کنیم، برآورد نقطه‌ای و بازه‌ای شما برای نرخ مشارکت دانشجویان بر اساس نمونه‌ای با اندازه‌ی ۲۰ چقدر است؟ پاسخ شما باید برای حداقل ۹۸ درصد جمعیت معتبر باشد.
